```python
from IPython.display import display, HTML, Image

import numpy as np
import pandas as pd
from pandas import Series
import re


def df_html(df, escape=[], formatters={}):
    pd.options.display.max_colwidth = 999999
    from itertools import repeat
    import cgi
    formatters = dict(zip(
        df.columns.values.tolist(),
        repeat(lambda x: cgi.escape(str(x)))
    ) + zip(
        escape,
        repeat(lambda x: x)
    ) + formatters.items())
    return HTML(df.to_html(escape=False, formatters=formatters))


def df_clean_columns(df):
    import string
    df.columns = df.columns.map(lambda s: filter(lambda x: x in string.printable, s))
```

```python
df = pd.read_csv("/Users/munro/Desktop/switches.csv")
df_clean_columns(df)

# clean up and filter misisng pricing
df['Unit Price (USD)'] = df['Unit Price (USD)'].convert_objects(convert_numeric=True)
df = df[df['Unit Price (USD)'].notnull()]

# parse circuits
circuit_size = ['Count', 'Inputs', 'Outputs']
df[circuit_size] = df['Circuit'].map(
    lambda x: re.search(r'(\d+)\s*x\s*(\d+):(\d+)', x).groups()
).apply(lambda loc: Series(loc, index=circuit_size)).astype(int)

# make fields interactive
df['Image'] = df['Image'].map(lambda x: '<img width="50" height="50" src="{}" />'.format(x))
df['Digi-Key Part Number'] = df.apply(lambda x: '<a href="{}">{}</a>'.format(x['"Datasheets"'], x['Digi-Key Part Number']), axis=1)

# trim, sort, and display
df = df[['Image', 'Digi-Key Part Number', 'Manufacturer', 'Description'] + circuit_size + ['Unit Price (USD)']]
df_html(df.sort(['Unit Price (USD)']).head(), escape=['Image', 'Digi-Key Part Number'])
```

```python
df['Switches'] = df['Count'] * df['Inputs'] * df['Outputs']
df['$/Switch'] = df['Unit Price (USD)'] / df['Switches']

df_html(df.sort(['$/Switch']).head(), escape=['Image', 'Digi-Key Part Number'])
```

```python
from IPython.display import display, HTML, Image

import pydot


class Thing(object):
    _name = None
    
    @property
    def name(self):
        import random
        if not self._name:
            self._name = 'n_{}'.format(str(random.random())[2:])
        return self._name
    
    def name_n(self, *args):
        return '{}_{}'.format(self.name, '_'.join([str(x) for x in args]))

    def my_sub(self, *args):
        return '_'.join([str(x) for x in args])

    def sub(self, *args):
        return '{}:{}'.format(self.name, self.my_sub(*args))

    
class Header(Thing):
    def __init__(self, cols, rows=1):
        self.cols = cols
        self.rows = rows
    
    @property
    def inputs(self):
        return self.cols * self.rows
    outputs = inputs
    
    def to_graph(self, g):
        g.add_node(pydot.Node(
            name=self.name,
            shape='record',
            label='|'.join([ 
                '{' + '|'.join(['<{}> {}'.format(self.my_sub('p', c + r * self.cols), c + r * self.cols) for r in range(0, self.rows)]) + '}'
                for c in range(0, self.cols)
            ])
        ))


class EthernetMux(Thing):
    def __init__(self, inputs):
        import math
        assert inputs > 0
        self.inputs = int(math.ceil(inputs / 2.0) * 2)

    @property
    def outputs(self):
        return self.inputs / 2

    def to_graph(self, g):
        for x in range(0, self.inputs):
            g.add_node(pydot.Node(self.name_n('in', x), shape='circle', label='', width='.15'))
            g.add_edge(pydot.Edge(
                self.name_n('in', x),
                self.name_n('out', int(x / 2)),
                dir='none'
            ))
        for x in range(0, self.outputs):
            g.add_node(pydot.Node(self.name_n('out', x), shape='circle', label='', width='.15'))

class ChainMux(Thing):
    def __init__(self, inputs, MuxClass):
        assert inputs > 0
        self.start = MuxClass(inputs=inputs)
        self.MuxClass = MuxClass
        
        def chain(last_mux):
            pass
            
    
    @property
    def inputs(self):
        return self.start.inputs

def display_graph(g):
    g.write_png('lda_graph.png', prog='dot')
    return Image('lda_graph.png')


# for real
g = pydot.Dot(shape='circle')

header = Header(cols=36, rows=1)
header.to_graph(g)

mux = EthernetMux(inputs=36)
mux.to_graph(g)

for x in range(0, min(mux.inputs, header.inputs)):
    g.add_edge(pydot.Edge(
        header.sub('p', x),
        mux.name_n('in', x),
        dir='none'
    ))

g.write_png('lda_graph.png', prog='dot')


for x in range(0, 5):
    next_mux = EthernetMux(inputs=mux.outputs)
    next_mux.to_graph(g)

    for x in range(0, mux.outputs):
        g.add_edge(pydot.Edge(
            mux.name_n('out', x),
            next_mux.name_n('in', x),
            dir='none'
        ))
    
    mux = next_mux


    
# print(g.to_string())
display_graph(g)
```
